```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load datasets
athletes=pd.read_csv('C:/Users/DELL/Downloads/athlete_events.csv.zip')
regions=pd.read_csv('C:/Users/DELL/Downloads/noc_regions.csv')
```

```python
regions.head()
```

```python
#join dataframes
athletes_df=athletes.merge(regions, how='left', on='NOC')
athletes_df.head()
```

```python
# col names consistent
#rename
athletes_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True);
athletes_df.head()
```

```python
#top countrys participate
top_10_countries=athletes_df.Team.value_counts(ascending=False).head(10)
top_10_countries
```

```python
#plot for 10 countries
plt.figure(figsize=(12,6))
plt.title('participation by countries');
sns.barplot(x=top_10_countries.index,y=top_10_countries,palatte='Set2');
```

```python
#age distribution
plt.figure(figsize=(12,6))
plt.title("age dist")
plt.xlabel('age')
plt.ylabel('participants')
plt.hist(athletes_df.Age,bins=np.arange(10,80,2),color='red',edgecolor='blue');
```

```python
winter_sports=athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
#male female participation
gender_counts=athletes_df.Sex.value_counts()
gender_counts
```

```python
#pie plot for male and female
plt.figure(figsize=(10,2))
plt.title('gender distribution')
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150,shadow=True);
```

```python
womenOlympics=athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')]
sns.set(style="darkgrid")
plt.figure(figsize=(10,20))
sns.countplot(x='Year',data=womenOlympics,palette="Spectral")
plt.title('women partipants')
```
